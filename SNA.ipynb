# Importing data from a csv file 

```python
#NetworkX includes many graph generator functions and facilities to read and write graphs in many formats.
import networkx as nx
import pandas as pd
#from google.colab import drive
#drive.mount('/content/drive')
path="/content/dolphins conn.csv"
data=pd.read_csv(path)
DG=nx.DiGraph()
G=nx.Graph() #creating a graph


for i in range(len(data)):
    DG.add_edge(data['_source'][i],data['_target'][i])
    G.add_edge(data['_source'][i],data['_target'][i])

nx.draw_spring(G,with_labels=True)
```

# Graph Metrics

Degree,Indegree and Outdegree 

```python
degree= G.degree(list(G.nodes))
indegree=DG.in_degree(list(DG.nodes))
outdegree=DG.out_degree(list(DG.nodes))
print("Degree: \n",degree,"\n\n\nIndegree:\n",sorted(indegree),"\n\n\nOutdegree:\n",sorted(outdegree))

color_map=[]

for node in G:
    if node!=12 and node!=35:
        color_map.append('blue')
    else: 
        color_map.append('green')  

nx.draw_spring(G,node_color=color_map,with_labels=True,font_color='white')
```

Density

```python
density=nx.density(DG)
density
#The density is the ratio of edges present in a graph divided by the maximum possible edges. 
```

Clustering Coefficient

```python
nx.average_clustering(DG)
#The average clustering coefficient of a graph G is the mean of local clusterings. 
#This function finds an approximate average clustering coefficient for G by repeating n times (defined in trials) the following experiment:
# choose a node at random, choose two of its neighbors at random, and check if they are connected.
```

Triadic Closure 

```python
nx.transitivity(G)
#Graph Transitivity is the overall probability for the network to have adjacent nodes interconnected, 
#thus revealing the existence of tightly connected communities (or clusters, subgroups, cliques).
```

 Betweenness centrality (looks at all the shortest paths that pass through a particular node ) (Brokers)

```python
print(nx.betweenness_centrality(DG))
#In graph theory, betweenness centrality (or "betweeness centrality") is a measure of centrality in a graph based on shortest paths.
# Betweeness centrality of v =Total number of shortest paths between node u and w that pass through v /Total number of shortest paths between node u and w 
```

Eigen vector centrality

Closeness Centrality

```python
print(nx.closeness_centrality(G))
# In a connected graph, closeness centrality (or closeness) of a node is a measure of centrality in a network, 
#calculated as the reciprocal of the sum of the length of the shortest paths between the node and all other nodes in the graph.
# Thus, the more central a node is, the closer it is to all other nodes
```

Strongly and Weakly connected components

```python
# A strongly connected component is a directed subgraph where nodes are connected in such a way that every vertex/node is reachable
# from other vertices/nodes of that subgraph.

#a weakly connected component is a directed subgraph that is unreachable from other nodes/vertices of a graph or subgraph.
print("Strongly Connected Component: ",nx.is_strongly_connected(DG))
print("Weakly Connected Component: ",nx.is_weakly_connected(DG))
```

# Algorithms

Shortest Path

```python
print(nx.shortest_path(G)[2])
```

Minimum Spanning Tree

```python
#A minimum spanning tree is a spanning tree in which the sum of the weight of the edges is as minimum as possible.
from networkx.algorithms import tree
mst=list(tree.minimum_spanning_edges(G, algorithm='prim', data=False))
MST=nx.Graph()
MST.add_edges_from(mst)
nx.draw(MST,with_labels=True)
len(G.edges)-len(MST.edges)
```

```python
print(nx.pagerank(DG,alpha=0.9))
```

# Clustering or Forming Communities

Girvan Newman Algorithm. 

```python
# The Girvan-Newman algorithm for the detection and analysis of community structure relies on the iterative 
# elimination of edges that have the highest number of shortest paths between nodes passing through them.
from networkx.algorithms import community
c=list(community.girvan_newman(G))
c[1]
```
